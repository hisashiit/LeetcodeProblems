```python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        vectorAll = [''] * len(str2)
        biggestString = ''
        
        for i in range(1, len(str2) + 1):
            candidate = str2[:i]
            if len(str1) % len(candidate) == 0 and len(str2) % len(candidate) == 0:
                times1 = int (len(str1) / len(candidate))
                times2 = int (len(str2) / len(candidate))
                if candidate * times1 == str1 and candidate * times2 == str2:
                    vectorAll[i - 1] = candidate

        biggestLen = 0
        for val in vectorAll:
            if len(val) > biggestLen:
                biggestLen = len(val)
                biggestString = val
                
        return biggestString
```
