```python
class Solution:
    def distributeCandies(self, n: int, limit: int) -> int:
        # Mata repetido
        def matarepetido(a: list):
            b = [sorted(x) for x in a]
            c = []
            for item in b:
                if item not in c:
                    c.append(item)
            return c

        # Conta permuta
        def contapermutacoes(a: list):
            count = 0
            for c in a:
                if c[0] == c[1] == c[2]:
                    count += 1
                elif c[0] == c[1] or c[1] == c[2] or c[0] == c[2]:
                    count += 3
                else:
                    count += 6
            return count

        # Gera tudo que é possivel
        list_kids = []
        for a in range(0, min(limit, n) + 1):
            for b in range(0, min(limit, n - a) + 1):
                c = n - a - b
                if 0 <= c <= limit:
                    list_kids.append([a, b, c])

        # Mata repete
        list_without_permut = matarepetido(list_kids)

        # Conta repetido
        return contapermutacoes(list_without_permut)
```
